```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.nytimes.com/"
r = requests.get(url)
```

```python
r.content
```

```python
body = r.content
```

```python
soup = BeautifulSoup(body)
soup.prettify()
```

```python
title_unicode = soup.find("title").contents[0]
title_unicode
```

```python
# from unicode to utf-8
title_unicode.encode('utf-8')
```
