```python
with open('dae_l1.yaml', 'r') as f:
    layer1_yaml = f.read()
print layer1_yaml
```

```python
hyper_params_l1 = {'train_stop' : 50000,
                   'batch_size' : 100,
                   'monitoring_batches' : 5,
                   'nhid' : 500,
                   'max_epochs' : 10,
                   'save_path' : '.'}
layer1_yaml = layer1_yaml % (hyper_params_l1)
print layer1_yaml
```

```python
train = yaml_parse.load(layer1_yaml)
train.main_loop()
```
